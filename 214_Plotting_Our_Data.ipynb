```python
#Plotting data from pandas

import pandas as pd
transactions = pd.read_excel("grocery_database.xlsx", sheet_name = "transactions" )
customer_details = pd.read_excel("grocery_database.xlsx", sheet_name = "customer_details" )
product_areas = pd.read_excel("grocery_database.xlsx", sheet_name = "product_areas" )
```

```python
#Default plot is with lines
#However, this chart type may not be useful
customer_details.plot()
```

```python
daily_sales_summary = transactions.groupby("transaction_date")[["sales_cost", "num_items"]].sum().reset_index()
```

```python
#Total sales and items by date
daily_sales_summary
```

```python
daily_sales_summary["sales_cost"].plot()
#Need to fix x-axis to be date
```

```python
#X-axis, Y-axis
daily_sales_summary.plot(x = "transaction_date", y = "sales_cost")
```

```python
#Kind can be used here to change chart type - Scatter
daily_sales_summary.plot(x = "transaction_date", y = "sales_cost",  kind= "scatter")
```

```python
#Kind can be used here to change chart type - Scatter
daily_sales_summary.plot(x = "num_items", y = "sales_cost",  kind= "scatter")
#High correlation between num_items and sales_cost
```

```python
#Box plot - focused on one variable...how values are distributed from low to high values
#Kind can be used here to change chart type - Scatter
daily_sales_summary.plot( y = "sales_cost",  kind= "box")
```

```python
#50% of data points reside in the box itself - between 25th and 75th
#The min is 5% (lowest line) and max is 95% (highest line)
#Circles are outliers
daily_sales_summary["sales_cost"].describe()
```

```python
#Histogram - vertical bars that represent frequency of data points
daily_sales_summary.plot( y = "sales_cost",  kind= "hist")
#Frequency default on y-axis is 10 point increments
#First bar says - 10 days in which sales feel between 5500-6800 dollars or so
#We may want more bins here
```

```python
daily_sales_summary.plot( y = "sales_cost",  kind= "hist", bins=25)
#Now we have further granularity here
```

```python
#Bar plot useful for visualizing different scales
product_areas.plot(kind = "bar", y="profit_margin", x = "product_area_name")
#Non-food area has highest profit margin at nearly 25%
```
